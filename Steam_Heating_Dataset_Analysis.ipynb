```python
def perc_missing(df):
    '''prints out columns with missing values with its %'''
    for col in df.columns:
        pct = df[col].isna().mean() * 100
        if (pct != 0):
            print('{} => {}%'.format(col, round(pct, 2)))
    
perc_missing(data)
```

```python
data.describe()
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], infer_datetime_format=True)
data.info()
```

```python
new_df = data
new_df.head()
```

```python
new_df.drop('Temperature (F)', axis=1, inplace=True)
new_df.head()
```

```python
new_df.info()
```

```python
new_df.bfill(inplace=True)
new_df.isnull().sum()
```

```python
new_df = new_df.set_index('timestamp')
new_df.head()
```

```python
new_df['Month'] =  new_df.index.month
new_df['Weekday Name'] =  new_df.index.weekday_name
new_df['Hour'] =  new_df.index.hour
new_df.sample(5, random_state=0)
```

```python
new_df['Day'] =  new_df.index.day
new_df.sample(5, random_state=0)
```

```python
DF = new_df
DF.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
cols_plot = ['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)']

axes = DF[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 21), subplots=True)

for ax in axes:
    ax.set_ylabel('Demand')
```

```python
fig, axes = plt.subplots(5, 1, figsize=(21, 21), sharex=True)

for name, ax in zip(['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)'], axes):
    sns.boxplot(data=DF, x='Month', y=name, ax=ax, order=[10,11,12,1])
    ax.set_ylabel('Demand')
    ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```

```python
fig, axes = plt.subplots(5, 1, figsize=(21, 21), sharex=True)

for name, ax in zip(['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)'], axes):
    sns.boxplot(data=DF[DF['Month']==12], x='Day', y=name, ax=ax)
    ax.set_ylabel('Demand')
    ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```

```python
data_columns = ['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)']

# Resample to Daily frequency, aggregating with mean
DF_daily = DF[data_columns].resample('D').mean()
DF_daily.head(3)
```

```python
fig, axes = plt.subplots(5, 1, figsize=(11, 21), sharex=True)

for name, ax in zip(['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)'], axes):
    DF_daily.loc['2017-12', name].plot(ax=ax)
    ax.set_ylabel('Demand')
    ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```

```python
data_columns = ['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)']

# Resample to monthly frequency, aggregating with mean
DF_monthly = DF[data_columns].resample('M').mean()
DF_monthly.head(3)
```

```python
fig, axes = plt.subplots(5, 1, figsize=(11, 21), sharex=True)

for name, ax in zip(['Steam Demand (klbs/hr)', 'AHU1 (kW)', 'AHU2 (kW)', 'AHU3 (kW)', 'AHU4 (kW)'], axes):
    DF_monthly.loc[:, name].plot(ax=ax, marker='o')
    ax.set_ylabel('Demand')
    ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```
